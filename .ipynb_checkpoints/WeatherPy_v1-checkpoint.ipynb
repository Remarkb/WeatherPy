* As shown in the first figure, the weather becomes warmer as one approaches the equator at zero degrees latitude. Due to the tilt of the Earth and opposing seasons in the northern and southern hemispheres, we see warmer temperatures in the latitudes below the equator at this time of year and much colder temperatures in the latitudes above the equator.   
* We see a large concentration of cities at 100 percent humidity, particularly near the equator and below.  This would indicate precipitation in the warmer southern hemispere at this time.
* While there doesn't seem to be a strong relationship between latitude and wind speed, we do see a greater concetration of higher wind speeds in the northern hemisphere.  This may be due in part to the larger temperature gradients that occur during the winter season.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time

#Import API key
from wxkey import wxkey

#Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Lists for response results
wx_data = {"City": [], "Cloudiness": [], "Humidity": [], 
                "Lat": [], "Lon": [], "Temperature": [], "Wind": []}


url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
            
query_url = f"{url}appid={wxkey}&units={units}&q="

count = 1

for city in cities:
    print(f" Calling record: {count}")
    city_wx_response = requests.get(query_url + city).json()
    #wx_data = json.dumps(city_wx_response, indent = 4, sort_keys=True)
    city_number = city_wx_response.get("id")
    
    if city_wx_response.get("id"):
        print(f"  Log data for city: {city}, City ID: {city_number}")
        
        #city.append(city_wx_response["name"])
        wx_data["City"].append(city_wx_response["name"])
        wx_data["Cloudiness"].append(city_wx_response["clouds"]["all"])
        wx_data["Humidity"].append(city_wx_response["main"]["humidity"])
        wx_data["Lat"].append(city_wx_response["coord"]["lat"])
        wx_data["Lon"].append(city_wx_response["coord"]["lon"])
        wx_data["Temperature"].append(city_wx_response["main"]["temp"])
        wx_data["Wind"].append(city_wx_response["wind"]["speed"])
        time.sleep(1.05)
        count = count + 1
    else:
        print(f"  No record found for {city}... Skipping.")
        time.sleep(1.05)      
    #print(wx_data)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
wx_data_df = pd.DataFrame(wx_data)

wx_data_df.head()
```

```python
wx_data_df.to_csv("CityWxData.csv", index=False, header=True)
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
plt.scatter(wx_data_df["Lat"],wx_data_df["Temperature"], marker="o", alpha=0.7, label="Cities", edgecolors="black",)

plt.title("City Latitude vs Temperature")
plt.xlabel("Latitude")
plt.ylabel("Temperature (Fahrenheit)")
plt.grid(True)


# Save Figure
plt.savefig("CityLatvTemp.png")

# Show Plot
plt.show()
```
